```python
import json
```

```python
dict_matching_res = json.load(open("../datasets/covid/dict-matching-res.txt", 'r'))
labels = json.load(open("../datasets/covid/labels.txt", 'r'))
labels = set(labels["LABEL"])
```

```python
# TRAIN_DATA = [(text, {"entities": [(start, end, label), ...]})]

def generate_entities(doc, title_offset, abs_offset):
    entities = doc['entities']
    entities_dict = {}
    entities_list = []
    for ent in entities:
        if ent['type'] in labels:
            start = ent['start']
            end = ent['end']
            label = ent['type']
            entities_list.append((start, end, label))
#             if end <= title_offset:
#                 entities_list.append((start, end, label))
#             elif start >= title_offset and end < abs_offset:
#                 entities_list.append((start, end, label))
#             elif start >= abs_offset:
#                 entities_list.append((start+2, end+2, label))
    entities_dict['entities'] = entities_list
    return entities_dict
```

```python
def generate_one_doc(doc):
    title_offset = len(doc['title'])
    abs_offset = len(doc['abstract']) + title_offset
    text = doc['title'] + ' ' + doc['abstract'] + ' ' + doc['body']
    entities_dict = generate_entities(doc, title_offset, abs_offset)
    return (text, entities_dict)
```

```python
def generate_training_data(docs):
    res = []
    for doc in docs:
        res.append(generate_one_doc(doc))
    return res
```
